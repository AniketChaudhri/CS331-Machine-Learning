#### (Perceptron) Using the make blobs function in sklearn generate a dataset of 100 points with two classes. Write the perceptron algorithm by scratch and show the intermediate hyperplanes generated.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_blobs, make_circles
```

```python
# generate 2 classes sample data of size 100
x, y = make_blobs(n_samples=100, centers=2, n_features=2, random_state=2)
```

```python
# plot the data
plt.scatter(x[:, 0], x[:, 1], c=y, cmap='rainbow')
```

```python
# Augment the data with a column of ones
x = np.hstack((np.ones((x.shape[0], 1)), x))
```

```python
# write perceptron algorithm
def perceptron(x, y, lr=1, epochs=100, plot=True, print_accuracy=True):
    w = np.random.randn(x.shape[1])
    while True:
        for i in range(len(x)):
            curr = np.dot(x[i], w)
            if curr > 0 and y[i] == 0:
                w -= lr * x[i]
            elif curr < 0 and y[i] == 1:
                w += lr * x[i]
        epochs-=1

        # isCorrectlyClassified = True
        # for i in range(len(x)):
        #     curr = np.dot(x[i], w)
        #     if curr > 0 and y[i] == 0:
        #         isCorrectlyClassified = False
        #     elif curr < 0 and y[i] == 1:
        #         isCorrectlyClassified = False
        
        # if epochs:
        #     # plot the line and points
        #     plt.scatter(x[:, 1], x[:, 2], c=y, cmap='rainbow')
        #     x1 = np.linspace(-10, 10, 100)
        #     x2 = -(w[0] + w[1] * x1) / w[2]
        #     plt.plot(x1, x2, 'k-')
        #     plt.show()

        # if isCorrectlyClassified or epochs == 0:
        #     break
        incorrectlyClassified = 0
        for i in range(len(x)):
            curr = np.dot(x[i], w)
            if curr > 0 and y[i] == 0:
                incorrectlyClassified += 1
            elif curr < 0 and y[i] == 1:
                incorrectlyClassified += 1

        if print_accuracy:
            # print accuracy
            print("Accuracy: ", (len(x) - incorrectlyClassified) / len(x) * 100, "%")
        if plot:
            # plot the line and points
            plt.scatter(x[:, 1], x[:, 2], c=y, cmap='rainbow')
            x1 = np.linspace(-10, 10, 100)
            x2 = -(w[0] + w[1] * x1) / w[2]
            plt.plot(x1, x2, 'k-')
            plt.show()

        if not incorrectlyClassified or epochs == 0:
            print("Accuracy: ", (len(x) - incorrectlyClassified) / len(x) * 100, "%")
            return w
            break
```

```python
perceptron(x, y)
```

#### (Perceptron) Repeat the above exercise with batch gradients instead. Step size might need to be adjusted for convergenence.

```python
# write perceptron algorithm for batch gradient descent
def perceptron_batch(x, y, lr=1, epochs=100, batch_size=10, plot=True, print_accuracy=True):
    # initialise w with random values
    w = np.random.randn(x.shape[1])
    while True:
        # generate random batch
        batch = np.random.randint(0, len(x), batch_size)
        for i in batch:
            curr = np.dot(x[i], w)
            if curr > 0 and y[i] == 0:
                w -= lr * x[i]
            elif curr < 0 and y[i] == 1:
                w += lr * x[i]
        epochs-=1

        incorrectlyClassified = 0
        for i in range(len(x)):
            curr = np.dot(x[i], w)
            if curr > 0 and y[i] == 0:
                incorrectlyClassified += 1
            elif curr < 0 and y[i] == 1:
                incorrectlyClassified += 1

        if print_accuracy:
            # print accuracy
            print("Accuracy: ", (len(x) - incorrectlyClassified) / len(x) * 100, "%")
        if plot:
            # plot the line and points
            plt.scatter(x[:, 1], x[:, 2], c=y, cmap='rainbow')
            x1 = np.linspace(-10, 10, 100)
            x2 = -(w[0] + w[1] * x1) / w[2]
            plt.plot(x1, x2, 'k-')
            plt.show()
    
        if not incorrectlyClassified or epochs == 0:
            print("Accuracy: ", (len(x) - incorrectlyClassified) / len(x) * 100, "%")
            return w
            break

        
```

```python
def perceptron_predict(x, w):
    return np.dot(x, w)
```

```python
def perceptron_accuracy(x, y, w):
    correctlyClassified = 0
    for i in range(len(x)):
        curr = np.dot(x[i], w)
        if curr < 0 and y[i] == 0:
            correctlyClassified += 1
        elif curr > 0 and y[i] == 1:
            correctlyClassified += 1
    return (correctlyClassified / len(x)) * 100
```
